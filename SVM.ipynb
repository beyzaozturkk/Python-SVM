**Destek Vektör Makinleri**


Digits veri seti, makine öğrenimi ve özellikle de görüntü tanıma alanında sıkça kullanılan temel bir veri setidir. Bu set, el yazısı rakamların (0-9) farklı örneklerden oluşan siyah-beyaz görüntülerini içerir. Her görüntü, küçük bir matrisle temsil edilir ve bu matristeki her bir eleman, ilgili pikselin yoğunluğunu (siyah veya beyaz) gösterir.

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
import matplotlib.pyplot as plt
```

```python
fig,axes=plt.subplots(nrows=2,ncols=5,figsize=(10,5),subplot_kw={'xticks':[],'yticks':[]})
for i,ax in enumerate(axes.flat):
  ax.imshow(digits.images[i],cmap="binary",interpolation="nearest")
  ax.set_title(digits.target[i])
plt.show
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=digits.data
Y=digits.target
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVC
```

```python
svc_clf=SVC(kernel="linear",random_state=0)
svc_clf.fit(X_train,Y_train)
```

```python
y_pred=svc_clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,y_pred))
```
